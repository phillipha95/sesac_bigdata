```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.describe()
```

```python
def draw_distplot(col_name):
    class_names = ['first', 'second', 'third']
    plt.figure(figsize=(12,5))
    for i in range(0,3):    
        plt.subplot(1,3,i+1)
        sns.distplot(titanic[titanic.pclass == (i + 1)][col_name], axlabel = class_names[i])
    plt.tight_layout()
    plt.show()
```

```python
draw_distplot('fare')
```

```python
print('skew : ', round(titanic.fare.skew(), 2))   #왜도   
print('kurtosis : ', round(titanic.fare.kurt(), 2)) #첨도


# 왜도 : 분포의 비대칭도 
# 정규분포 = 0
# 왼쪽으로 치우침 왜도 >0
# 오른쪽으로 치우침 왜도 <0

# 첨도
# 확률분포의 뾰족한 정도
#  정규분포 첨도 0 >> 일반적인 형태 구글 검색 확인
#  위로 뾰쪽함 = 첨도 >0
# 아래쪽으로 뾰족함 = 첨도 <0
```

```python
from sklearn import preprocessing

# sklearn > 알고리즘의 모델과 데이터의 트레이닝 테스트를 할수 있는 패키지이다.
# 중요 하다.
# 표준화 or 정규화 작업을 진행하여 
```

```python
titanic['fare_log'] = preprocessing.scale(np.log(titanic.fare+1))
draw_distplot('fare_log')
```

```python
print('skew : ', round(titanic.fare_log.skew(), 2))
print('kurtosis : ', round(titanic.fare_log.kurt(), 2))
```

```python
from scipy import stats

fig = plt.figure(figsize = (12,8))

fig.add_subplot(1,2,1)
res = stats.probplot(titanic['fare'], plot=plt)

fig.add_subplot(1,2,2)
res = stats.probplot(titanic['fare_log'], plot=plt)
```
